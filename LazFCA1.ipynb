```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing, metrics
```

```python
def make_intent(example):
    attrib = []
    for j in range(len(example)):
        attrib.append(j)
        
    return set([str(i)+':'+str(k) for i,k in zip(attrib,example)])
```

```python
def intersect(example, context):
    intent = make_intent(example)
    intersection_res = []
    count = 0
    for i in context:
        intersection_x = intent&i
        if intersection_x:
            count = count + 1
            intersection_res.append(intersection_x)
    return [intersection_res, count]
```

```python
def aggregate1(plus_massiv,  plus_count, minus_massiv, minus_count):
    psum = sum([len(p) for p in plus_massiv])
    size_plus = len(plus_massiv)
    msum = sum([len(p) for p in minus_massiv])
    size_minus = len(minus_massiv)
    k = plus_count/psum
    z = minus_count/msum
    return z>=k
```

```python
def classifier1(X_train, X_test, y_train, y_test):
    X_tr = X_train.values.tolist()
    X_ts = X_test.values.tolist()
    y_tr = y_train.values.tolist()
    y_ts = y_test.values.tolist()

    plus_intent = [make_intent(O) for i, O in enumerate(X_tr) if y_tr[i]] 
    
    minus_intent = [make_intent(O) for i, O in enumerate(X_tr) if not y_tr[i]]
    
    results = {'tp': 0, 'fp': 0, 'tn': 0, 'fn': 0, 'contr': 0}
    
    for j, i in enumerate(X_ts):
        label = {'plus': False, 'minus': False}
        
        plus = intersect(i, plus_intent)
        minus = intersect(i, minus_intent)
        plus_massiv = plus[0]
        plus_count = plus[1]
        minus_massiv = minus[0]
        minus_count = minus[1]
        if aggregate1(plus_massiv,  plus_count, minus_massiv, minus_count):
            label['plus'] = True
        else:
            label['minus'] = True
        if label['plus'] and not label['minus']:
            if y_ts[j]:
                results['tn']+=1
            else: 
                results['fn']+=1
        elif label['minus'] and not label['plus']:
            if y_ts[j]:
                results['fp']+=1
            else: 
                results['tp']+=1
        else:
            results['contr']+=1
    return results
```

```python
X = df[[d for d in df.head() if d != 'target']]
Y = df['target']


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state = 1)
classif1 = classifier1(X_train, X_test, y_train, y_test)

```

```python
accur = (classif1['tp'] + classif1['tn']) / (classif1['tp'] + classif1['tn'] + classif1['fp'] + classif1['fn'])
precision = classif1['tp'] / (classif1['tp'] + classif1['fp'])

recall = classif1['tp'] / (classif1['tp'] + classif1['fn'])
print("Aggregation function:   ")
print(" ")
print("accuracy:  ", accur*100)
print("precission:  ", precision*100)
print("recall:   ", recall*100)
```
